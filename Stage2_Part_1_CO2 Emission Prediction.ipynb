# Data Analysis

```python
# Range of "Year"
df['Year'].unique()
# We have data from 2000 to 2019 which is different from the actual file which was supposed to contain the data  
# from 2000 to 2020, which shows many values of 2020 were null values.
```

Top 10 CO2 emitting countries in 2019 

```python
top_2019=df[df['Year']==2019][['Entity', 'Value_co2_emissions_kt_by_country']].sort_values(by=  'Value_co2_emissions_kt_by_country',ascending=False)
top_2019['Rank']= top_2019['Value_co2_emissions_kt_by_country'].rank(ascending=False)
top_2019.head(10)
```

Top 10 CO2 emitting countries in 2000

```python
top_2000=df[df['Year']==2000][['Entity', 'Value_co2_emissions_kt_by_country']].sort_values(by=  'Value_co2_emissions_kt_by_country',ascending=False)
top_2000['Rank']= top_2000['Value_co2_emissions_kt_by_country'].rank(ascending=False)
top_2000.head(10)
```

Checking the 2019 rank of the United Kingdom, Italy, and France. 

```python
top_2019[(top_2019['Entity']=='United Kingdom') |  (top_2019['Entity']=='Italy') | (top_2019['Entity']=='France')]
```

Checking the 2000 rank of the  Indonesia, Saudi Arabia and South Africa

```python
top_2000[(top_2000['Entity']== 'Indonesia') |(top_2000['Entity']== 'Saudi Arabia')|(top_2000['Entity']== 'South Africa' )]
```

We are going to Focus on these 6 countries along with India, China, United States, and Japan to understand how things had changed in these 19 year for them.

```python
nation_filter= ['South Africa','Indonesia','Saudi Arabia','United Kingdom','Italy','France','India','China','United States','Japan']
```

```python
nations= df[df['Entity'].isin(nation_filter)]
```

```python
#for i in nation_filter:
   # plt.figure(figsize=(5,5))
   # sns.lineplot(x= 'Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']== i])
   # plt.xlabel("{}--Year".format(i))
plt.figure(figsize=(18,18))
for i in nation_filter:
    sns.lineplot(x= 'Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']== i],label=i)
    


```

```python
# Correlation of Value_co2_emissions_kt_by_country with different features in nations df.
nations.corr()['Value_co2_emissions_kt_by_country'].sort_values(ascending=False).drop('Value_co2_emissions_kt_by_country')
```

```python
# Correlation of Value_co2_emissions_kt_by_country with different features in df (Which is world data).
df.corr()['Value_co2_emissions_kt_by_country'].sort_values(ascending= False).drop('Value_co2_emissions_kt_by_country')
```

```python
for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Electricity from fossil fuels (TWh)',data=nations[nations['Entity']==i],label='Electricity from fossil fuels (TWh)',color='green')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Electricity from renewables (TWh)',data=nations[nations['Entity']==i],label='Electricity from renewables (TWh))',color='green')
    plt.legend()
    
```

```python
# Let's see a realationship between renewable energy share and low carbon electricity for these countries with CO2 emissions. 
for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Renewable energy share in the total final energy consumption (%)',data=nations[nations['Entity']==i],label='Renewable share in the total energy consumption (%)',color='green')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Low-carbon electricity (% electricity)',data=nations[nations['Entity']==i],label='Low-carbon electricity (% electricity)',color='green')
    plt.legend()
    
```

```python
# Let's compare electricty from nuclear and access to clean fules for cooking with CO2 emissions.

for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Electricity from nuclear (TWh)',data=nations[nations['Entity']==i],label='Electricity from nuclear (TWh)',color='green')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Access to clean fuels for cooking',data=nations[nations['Entity']==i],label='Access to clean fuels for cooking',color='green')
    plt.legend()



```

```python
# GDP and gdp_per_capita vs CO2 emission 
for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='gdp_growth',data=nations[nations['Entity']==i],label='gdp_growth',color='green')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='gdp_per_capita',data=nations[nations['Entity']==i],label='gdp_per_capita',color='green')
    plt.legend()
```

```python
# Acess to electrcity and population density vs CO2 emission
for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Access to electricity (% of population)',data=nations[nations['Entity']==i],label='Access to electricity in %',color='green')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Density\\n(P/Km2)',data=nations[nations['Entity']==i],label='Population Density',color='green')
    plt.legend()
```

```python
# Energy intensity level of primary energy and Primary energy consumption per capita vs CO2 emission.
for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Energy intensity level of primary energy (MJ/$2017 PPP GDP)',data=nations[nations['Entity']==i],label='Energy intensity level of primary energy',color='red')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Primary energy consumption per capita (kWh/person)',data=nations[nations['Entity']==i],label='Primary energy consumption per capita',color='green')
    plt.legend()
```
