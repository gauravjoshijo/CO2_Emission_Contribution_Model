```python
# Creating a dataframe containing data for these 6 countries along with India, China, the United States, and Japan.
nation_filter= ['South Africa','Indonesia','Saudi Arabia','United Kingdom','Italy','France','India','China','United States','Japan']
```

```python
nations= df[df['Entity'].isin(nation_filter)]
```

```python
# Correlation heatmap of nations dataframe. 
plt.figure(figsize=(20,20))
sns.heatmap(nations.corr(),annot= True)
```

```python
# Comparison of trendlines of Co2 emissions, electricity from fossil fuels and electricity from renewables for nations dataframe.

for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Electricity from fossil fuels (TWh)',data=nations[nations['Entity']==i],label='Electricity from fossil fuels (TWh)',color='black')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Electricity from renewables (TWh)',data=nations[nations['Entity']==i],label='Electricity from renewables (TWh))',color='green')
    plt.legend()
    
```

Some general observations that can be made from above graphs. 


- Electricity from renewables has increased for all the selected countries.<br>

- The countries which have seen an overall decline in CO2 emission trend have somewhat seen a decrease in electricity from fossil fuels contribution. <br>

- The trend lines of CO2 emission of multiple countries seem to resemble the electricity from fossil fuels trend lines, specifically for the countries where the CO2 emission has declined.<br> 

- France is the only country in the list where the range of electricity from fossil fuels (35-65 TWh) is less than the range of electricity from renewables (60-110 TWh).

```python
# Comparison of trendlines of Co2 emissions, electricity from nuclear, and energy intensity level of primary energy.

for i in nation_filter:
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    sns.lineplot(x='Year',y='Value_co2_emissions_kt_by_country',data=nations[nations['Entity']==i],label='Value_co2_emissions_kt_by_country',color='red')
    plt.subplot(1,3,2)
    sns.lineplot(x='Year',y='Electricity from nuclear (TWh)',data=nations[nations['Entity']==i],label='Electricity from nuclear (TWh)',color='green')
    plt.title(i,loc='center')
    plt.subplot(1,3,3)
    sns.lineplot(x='Year',y='Energy intensity level of primary energy (MJ/$2017 PPP GDP)',data=nations[nations['Entity']==i],label='Energy intensity level of primary energy (MJ/$2017 PPP GDP)',color='blue')
    plt.legend()
```

Some general observations that can be made for nations dataframe from above graphs. 

- Saudi Arabia, Indonesia, and Italy do not use nuclear energy for electricity production. <br>
- Since the Fukushima nuclear disaster that happened in 2011, there has been a visible decline in Japan’s nuclear energy use for electricity.<br>  
- There is a decline in nuclear energy use for electricity for the United Kingdom and France whereas China and India have seen growth in it.<br>
- South Africa and the USA have seen ups and downs in the use of nuclear energy for electricity.<br> 
- There is no general relationship visible for electricity from nuclear energy for these countries with CO2 emission. It may have an impact on it, but from these trendlines it's hard to figure it out.<br>
- Except for Saudi Arabia, all other selected countries have seen decline in energy intensity level. 

**Note:** Energy intensity refers to the amount of energy used to produce a unit of economic output, usually measured in terms of GDP (Gross Domestic Product). It is an indicator that helps assess the efficiency of energy use in an economy. <br><br>
- **Energy Intensity Level = (Primary Energy Consumption) / (Economic Output or GDP)** 

```python
# Trendline comparison of Energy Intensity Level with gdp per capita.
for i in nation_filter:
    nation_data = nations[nations['Entity'] == i]
    
    fig = px.line(nation_data, x='Year', y='Value_co2_emissions_kt_by_country', title=f'{i} CO2 Emissions')
    fig.add_scatter(x=nation_data['Year'], y=nation_data['gdp_growth'], mode='lines', name='GDP Growth', yaxis='y2')
    
    fig.update_layout(
        yaxis=dict(title='CO2 Emissions (kt)'),
        yaxis2=dict(title='GDP Growth', overlaying='y', side='right'),
        xaxis=dict(title='Year'),
        legend=dict(x=0, y=1)
    )
    
    fig.show()

```

Following observations can be made from above graphs: 

- All the countries saw a dip in GDP growth in 2009 which can be attributed to the global financial crisis 2007-2008. <br>
- CO2 emissions also saw little dip in 2009 for some countries, but it's hard to say if it's directly and indirectly due to the dip in GDP growth. <br>
- There is no such evidence, at least from these trendlines, that GDP growth impacts CO2 emission to considerable impact. <br>
