# Feature Transformation

Goal of a model is to help in ranking countries on the basis of their predicted contribution in global CO2 emission. But from the above graphs it is understood that taking whole data will impact the accuracy of the results as there are many outliers present in the data. One method to solve the outlier issue is to remove the outliers and another is dividing the data into parts. From Value_co2_emissions_kt_by_country distribution it can be seen that most of the values are close to 0 i.e. they do not have a considerable contribution to global CO2 emissions. These countries can be removed as they do not have an impact on the overall percentage of CO2 emissions, but it should be done carefully as some parameters are required to filter out the data which do not impact much of the overall CO2 emissions. <br>

Also, it is beneficial for our prediction model to identify more patterns in countries CO2 emission behavior in order to divide data further into different groups corresponding to emission behavior. The methodology which will be adopted is **pareto analysis** which will help in finding the different contributions of different countries. <br>

Data will also be transformed in such a way that values in each feature represent a percentage contribution to the overall percentage corresponding to a year. Average of Value Co2 emissions for all the years will be taken in pareto analysis. 


```python
# Transforming values of features into percentage of their contribution to overall percentage for each year.
unique_years = df['Year'].unique()
cols= list(df.columns)
cols.remove('Entity')
cols.remove('Year')
for col in cols:
    total_by_year = df.groupby('Year')[col].sum()
    for year in unique_years:
        df.loc[df['Year'] == year, col] = df[df['Year'] == year][col] / total_by_year[year] * 100

    
```
